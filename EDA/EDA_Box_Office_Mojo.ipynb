```python
import pandas as pd
df = pd.read_csv("../Data/BoxOfficeMojo/box_office_mojo_cleaned.csv")
df.head()
```

```python
df['top10_gross'] = pd.to_numeric(df['top10_gross'].replace('[\$,]', '', regex=True), errors='coerce')
df = df.dropna(subset=['top10_gross'])
df['top10_gross'] = df['top10_gross'].astype('int')
df = df.sort_values(by=['year' , 'week_no'] , ascending= True)
df
```

```python
df[['weekend' , 'top10_gross']]
```

```python
import pandas as pd
import json
from datetime import datetime

# assuming df is your dataframe
data_list = []

for index, row in df.iterrows():
    date_str = row['weekend']
    # You will need a function here to convert date_str to a timestamp.
    # Assuming date_str is 'YYYY_MMM_wX' (like '1977_May_w4')
    year, month, week = date_str.split('_')
    month = datetime.strptime(month, "%b").month  # Convert three letter month abbreviation to month number
    day = int(week[-1]) * 7  # Convert week number to day number
    
    if day > 0:  # if day is not the first of the month, adjust it
        day = day - 6  # adjust day to the first day of the mentioned week
        
    timestamp = datetime(year=int(year), month=month, day=day).timestamp() * 1000  # to milliseconds
    value = row['top10_gross']
    
    data_list.append([timestamp, value])


```

```python
data_list[0:10]
```

```python
with open('bots_data.txt', 'w') as f:
    for item in data_list:
        f.write('%s\n' % item)
```

```python
with open('bots_data2.txt', 'w') as f:
    f.write("[\n")
    for i, item in enumerate(data_list):
        # if it is the last item, don't add a comma at the end
        if i == len(data_list) - 1:
            f.write("  %s\n" % item)
        else:
            f.write("  %s,\n" % item)
    f.write("]\n")
```

```python
# Replace . and \n in the 'genre' column
df_genre['genre'] = df_genre['genre'].str.replace('[.:\n]', '', regex=True)
```

```python
df_genre['genre'] = df_genre['genre'].replace('Science-Fiction', 'Sci-Fi')
```

```python
df_genre['genre'].value_counts()[0:20]
```

```python
keep_genres = ['Action', 'Comedy', 'Drama', 'Adventure', 'Animation', 'Thriller', 'Horror', 'Sci-Fi', 'Fantasy', 'Romance', 'Mystery', 'Musical', 'Western', 'Crime', 'Biography']
df_genre['genre'] = df_genre['genre'].where(df_genre['genre'].isin(keep_genres), 'Other')
df_genre['genre'].value_counts()
```

```python
genre_year = df_genre[['year' , 'genre']]
genre_year
```

```python
# Group by year and genre and count the occurrences
genre_counts = genre_year.groupby(['year', 'genre']).size().reset_index(name='count')

# Create the desired dictionary format
dataset = {}
for genre in genre_counts['genre'].unique():
    genre_data = genre_counts[genre_counts['genre'] == genre]
    dataset[genre] = dict(zip(genre_data['year'].astype(str), genre_data['count']))
```

```python
dataset
```

```python
import json

# Convert the dataset dictionary to a JSON string with double quotes
dataset_json = json.dumps(dataset, indent=4)  # The indent argument is optional; it just formats the output to be more readable

print(dataset_json)
```

```python
# Define the filename
filename = "dataset.txt"

# Write the JSON string to the file
with open(filename, 'w') as file:
    file.write(dataset_json)
```
